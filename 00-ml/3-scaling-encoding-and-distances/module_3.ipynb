```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({
    'h':[150,160,170,180,190],
    'w':[50,60,70,80,90]
})

mean=df.mean()
std=df.std()
print(mean)
print(std)
z=(df-mean)/std
z.round(2)
```

```python
z.plot(kind='bar')
plt.title("Z scaling features")
plt.xlabel("Index")
plt.ylabel("z value")
plt.show()
```
