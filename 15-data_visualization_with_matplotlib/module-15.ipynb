## Data Visualization with Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Line CHart

```python
hours=[2,3,4,1,7,9,5]
days=[1,2,3,4,5,6,7]
plt.plot(days,hours,marker='o',color='b',linestyle='--')
```

```python
hours=[2,3,4,1,7,9,5]
days=[1,2,3,4,5,6,7]
plt.ylabel('Hours Studied')
plt.xlabel('Days')
plt.plot(days,hours,marker='o',color='b',linestyle='--')
```

```python
hours1=[2,3,4,1,7,9,5]
hours2=[1,4,3,1,4,10,1]
days=[1,2,3,4,5,6,7]
plt.ylabel('Hours Studied')
plt.xlabel('Days')
plt.title('Study Hours Over a Week')
plt.plot(days,hours1,marker='o',color='b',linestyle='--')
plt.plot(days,hours2,marker='o',color='r',linestyle='--')
```

## 2D Line Plots from Pandas

```python
df=pd.read_csv('enrollment_data.csv')

df
```

```python
x=df['Year']
y1=df['Programming']
y2=df['Digital Marketing']

plt.plot(x,y1,marker='o',color='b',linestyle='-',label='Programming')
plt.xticks(x)
```

```python
plt.plot(x,y1,marker='o',color='b',linestyle='-',label='Programming')
plt.plot(x,y2,marker='o',color='r',linestyle='--',label='Digital Marketing')
plt.xticks(x)
plt.legend()
```

## Scatter plot

```python
students=pd.read_csv('student_IQdata.csv')

iqScore=students['IQ_Score']
chillingHours=students['Chilling_Hours']
studyHours=students['Study_Hour']
shoeSize=students['Shoe_Size']
students
```

```python
plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('COrelation between studty hours and IQ Score')
plt.scatter(studyHours,iqScore)
```

```python
#shoe size vs iq score
plt.xlabel('Shoe Size')
plt.ylabel('IQ Score')
plt.title('Correlation between Shoe Size and IQ Score')
plt.scatter(shoeSize,iqScore)
```

```python
#chilling hours vs iq score
plt.xlabel('Chilling Hours')
plt.ylabel('IQ Score')
plt.title('Correlation between chilling Hours and IQ Score')
plt.scatter(chillingHours,iqScore)
```
