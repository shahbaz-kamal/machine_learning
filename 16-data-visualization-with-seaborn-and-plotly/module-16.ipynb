```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
student=pd.read_csv('sns_data.csv')
student
```

## Lineplot using Seaborn

```python
sns.lineplot(data=student, x='week', y='attendance_rate',errorbar=None)
```

```python
sns.lineplot(data=student, x='week',y='attendance_rate',errorbar=None,hue='gender')
```

```python
sns.relplot(kind='line',data=student, x='week',y='test_score',errorbar=None, hue='gender')
```

```python
student=pd.read_csv('student_dataset_complete.csv')
student
```

## Facet plot using seaborn

```python
sns.scatterplot(data=student,x='study_hours', y='test_score', hue='gender')
```

```python
sns.relplot(kind='scatter',data=student,x='study_hours', y='test_score',col='gender')
```

```python
sns.relplot(kind='scatter',data=student,x='study_hours',y='test_score',col='gender',row='hostel')
```

## Histogram using seaborn

```python
# axes level
sns.histplot(data = student, x = 'attendance_rate',hue='gender')
```

## KDE Plot using seaborn

```python
sns.kdeplot(data=student,x='attendance_rate')
```

```python
tips_data=sns.load_dataset('tips')
tips_data
```

```python
sns.kdeplot(data=tips_data,x='total_bill',hue='sex',fill=True)
```

## COunt plot using seaborn

```python
sns.countplot(data=student,x='gender',hue='subject')
```

## Bar PLot

```python
sns.barplot(data=student,x='gender',y='Marks',errorbar=None,estimator=np.max)
```

## REglplot and Implot

```python
# axes level
sns.regplot(data=student, x='study_hours', y='test_score')
```

```python
# figure level

sns.lmplot(data=student, x='study_hours',y='test_score')
```

## pairplot and jointplot

```python
student_marks=student[['Marks','study_hours','attendance_rate','gender']]
student_marks
sns.pairplot(student_marks)
```

```python
# jointplot
sns.jointplot(data=student_marks,x='study_hours',y='Marks',kind='kde',hue='gender')
```
