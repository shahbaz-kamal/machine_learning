## filtering based on string

```python
import pandas as pd

data={
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva","Rahim","Karim","hola"],
    "City": ["New York", "Los Angeles", "Chicago new", "Houston", "Phoenix","Dhaka","Cumilla","Ne Los Angeles"],
    "Department":["HR", "Finance", "IT", "Marketing", "Sales","Sales","marketing","Hola"],
    "Salary":[70000, 80000, 90000, 75000, 85000,20000,22000,5000]
}

df=pd.DataFrame(data)

newCity=df.loc[df['City'].str.contains("New",case=False)]
print(df,"\n")
print(newCity)
```

```python
# starts with los


just_los=df.loc[df['City'].str.contains(r"^Los")]
# print("\n",just_los)
just_los
```

```python
#end with rk

end_with_rk=df.loc[df['City'].str.contains(r"rk$")]
print("\n",end_with_rk)
```

```python
# Name Starts with vowels

print(df.loc[df["Name"].str.contains(r"[AEIOU]")])

```

## Apply Functions

```python
#min max scaling

mn=df['Total Marks'].min()
mx=df['Total Marks'].max()

df['scaled Marks']=df["Total Marks"].apply(lambda x: (x-mn)/(mx-mn))
df
```

```python
# Grading system

def grading_systems(marks):
    if marks>=270:
        return 'A+'
    elif marks>=240:
        return 'A'
    elif marks>=210:
        return 'B+'
    elif marks>=180:
        return 'B'
    else:
        return 'C'

df["Grade"]=df['Total Marks'].apply(grading_systems)
df
```

```python
def marking_system(df):
    a=df['Data Structure Marks']*2
    b=df['Python Marks']*3
    c=df['Algorithm Marks']*4
    total=a+b+c
    return total

df['Exceptional Marks']=df.apply(marking_system,axis=1)
df
```
