## filtering based on string

```python
import pandas as pd

data={
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva","Rahim","Karim","hola"],
    "City": ["New York", "Los Angeles", "Chicago new", "Houston", "Phoenix","Dhaka","Cumilla","Ne Los Angeles"],
    "Department":["HR", "Finance", "IT", "Marketing", "Sales","Sales","marketing","Hola"],
    "Salary":[70000, 80000, 90000, 75000, 85000,20000,22000,5000]
}

df=pd.DataFrame(data)

newCity=df.loc[df['City'].str.contains("New",case=False)]
print(df,"\n")
print(newCity)
```

```python
# starts with los


just_los=df.loc[df['City'].str.contains(r"^Los")]
# print("\n",just_los)
just_los
```

```python
#end with rk

end_with_rk=df.loc[df['City'].str.contains(r"rk$")]
print("\n",end_with_rk)
```

```python
# Name Starts with vowels

print(df.loc[df["Name"].str.contains(r"[AEIOU]")])

```

## Handling Duplicate value

```python
data={
    "Name": ["Alice", "Bob", "Charlie", "Alice", "Bob","Rahim","Rahim","hola"],
    "City": ["New York", "Los Angeles", "Chicago new", "Los Angeles", "Los Angeles","Dhaka","Cumilla","Ne Los Angeles"],
    "Department":["HR", "Finance", "IT", "Marketing", "Finance","Sales","marketing","Hola"],
    "Salary":[70000, 80000, 90000, 75000, 80000,20000,20000,5000]
}
df=pd.DataFrame(data)
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(subset=['Name'])
```

```python
df.drop_duplicates(subset=['City'],keep='last')
```
